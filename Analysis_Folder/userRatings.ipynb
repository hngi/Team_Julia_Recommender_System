# Predicting User Ratings

Using data obtained from sql analysis of the lucid database we predict user ratings on articles read by guaging how the user interacted with the post.

```python
#importing dependencies

import numpy as np
import pandas as pd

#load notifications.csv

columns= ("id", "post_id", "comment_id", "comment", "sender_id", "user_id", "status", "action", "type", "timestamp", "updated_at")
notifications= pd.read_csv("notifications (1).csv", names= columns)

#convert to Dataframe
reactionData= pd.DataFrame(notifications)

reactionData.head()
```

```python
# analyis of reaction data

reactionData["action"].value_counts()
```

```python
#removing followed from action as it isn't strongly related to posts

indexNames = reactionData[ reactionData['action'] == "Followed"].index

reactionData.drop(indexNames, inplace= True)
```

```python
#dropping irrelevant columns

reactionData.drop(['id','comment_id', 'comment','type','updated_at', 'sender_id', 'status'], axis=1, inplace= True)

reactionData.head()
```

```python
#ranking action type

action_type = {'Love':5,'Like':4,'Commented': 3, 'Replied':2}

reactionData['Ratings'] = reactionData['action'].apply(lambda x: action_type[x])

reactionData.head()
```

```python
#reanrranging dataframe

userRatings = reactionData[['user_id','post_id','Ratings','timestamp']]

userRatings.head()
```

```python
#sorting values according to user_id

userRatings= userRatings.sort_values(by='user_id', axis= 0, ascending= True)

userRatings.head()
```

```python
userRatings.describe()
```

```python
#exporting userRatings to csv

userRatings.to_csv('userRatings.csv')
```
